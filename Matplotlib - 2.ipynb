```python
countries = ['India', 'UK', 'US', 'Japan']
covid_cases_total = [800, 450, 560, 90]
recovered_cases = [790, 410, 555, 87]

x = np.arange(len(countries))
  
plt.bar(x - 0.2, covid_cases_total, 0.4, label = 'Total')
plt.bar(x + 0.2, recovered_cases, 0.4, label = 'Recovered')
plt.xticks(x, countries)

plt.legend()
```

```python
countries = ['India', 'UK', 'US', 'Japan']
covid_cases_total = np.array([800, 450, 560, 90])
recovered_cases = np.array([790, 410, 555, 87])
deaths = covid_cases_total - recovered_cases

x = np.arange(len(countries))
# x = np.array([0,1,2,3])
  
plt.bar(x - 0.2, covid_cases_total, 0.2, label = 'Total')
plt.bar(x, recovered_cases, 0.2, label = 'Recovered')
plt.bar(x + 0.2, deaths, 0.2, label = 'Deaths')
plt.xticks(x, countries)

plt.legend()
```

```python
exp = np.array([1, 5, 8, 10, 15, 20])
sal = np.array([10000, 20000, 25000, 30000, 35000, 45000])

fig = plt.figure()

plt.subplot(2,1,1)
plt.bar(exp, sal)
plt.xticks(np.arange(0, 25, 5))
plt.yticks(np.arange(0, 50000, 10000))
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('My bar plot')

plt.subplot(2,1,2)
plt.scatter(exp, sal)
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.xticks(np.arange(0, 25, 5))
plt.yticks(np.arange(0, 50000, 10000))
plt.title('My Scatter plot')
fig.tight_layout()

plt.show()

```
