```python
fig = plt.figure()
fig.add_subplot(2,2,1) # rows, cols, position
plt.plot(np.array([1,2,3,4]))
fig.add_subplot(2,2,2)
fig.add_subplot(2,2,3)
fig.add_subplot(2,2,4)
```

```python
fig = plt.figure()
fig.add_subplot(2,2,1) # rows, cols, position
plt.plot(np.array([1,2,3,4]))
fig.add_subplot(2,2,2)
plt.plot(np.array([11,0,-4,2]))
fig.add_subplot(2,2,3)
plt.plot(np.array([1,1,1,1]))
fig.add_subplot(2,2,4)
plt.plot(np.array([6,5,0,10,4]))
```

```python
fig = plt.figure()
fig.add_subplot(2,2,1) # rows, cols, position
plt.plot(np.array([100,150,200,250]), np.array([1,2,3,4]))
fig.add_subplot(2,2,2)
plt.plot(np.array(['a','b','c','d']),np.array([11,0,-4,2]))
fig.add_subplot(2,2,3)
plt.plot(np.array([1,1,1,1]))
fig.add_subplot(2,2,4)
plt.plot(np.array([6,5,0,10,4]))
```
